```python
import pandas as pd

# '주문일자'가 문자열 형식이라면 날짜 형식으로 변환합니다.
df['주문일자'] = pd.to_datetime(df['주문일자'])
# '할부기간' 컬럼의 NaN 값을 1로 채우기하고 정수형으로 변환
df['할부개월수'] = df['할부개월수'].fillna(1).astype(int)

# 날짜 및 순수익을 저장할 빈 리스트를 생성합니다.
list1 = []

# DataFrame을 순회하며 날짜 및 순수익 정보를 리스트에 추가합니다.
for index, row in df.iterrows():
    for i in range(row['할부개월수']):
        if row['할부개월수'] == 1:
            list1.append({
                '날짜': (row['주문일자'] + pd.DateOffset(months=i)).strftime('%m'),
                '순수익': row['판매금액']
            })
        else :
            list1.append({
                '날짜': (row['주문일자'] + pd.DateOffset(months=i+1)).strftime('%m'),
                '순수익': row['판매금액'] / row['할부개월수']
            })

# 리스트를 기반으로 DataFrame을 생성합니다.
list1_data = pd.DataFrame(list1)

# '날짜'를 기준으로 그룹화하여 순수익을 합산합니다.
monthly_profit = list1_data.groupby('날짜').sum().reset_index()

# 결과를 출력합니다.
monthly_profit
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='AppleGothic'
plt.figure(figsize=(12, 6))
sns.barplot(x='날짜', y='순수익', data=monthly_profit)
plt.xlabel("월")
plt.title(f'월별 순수익')
plt.show()
```
