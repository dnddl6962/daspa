```python
# ' + 포', '-'는 결측치로 생각해서 삭제할 계획
df1 = df1.drop(df1[df1['결제방법'] == ' + 포'].index)
df1 = df1.drop(df1[df1['결제방법'] == '-'].index)
```

```python
df1.groupby(['결제방법']).agg({'판매금액': 'sum'})
```

```python
df1['결제방법'] = df1['결제방법'].astype(str) # 내용 문자열로 변환
df1.loc[df1.query('결제방법.str.startswith("신")').index, '결제방법'] = '신용카드'
df1.loc[df1.query('결제방법.str.startswith("적")').index, '결제방법'] = '적립금'
df1.loc[df1.query('결제방법.str.startswith("현")').index, '결제방법'] = '현금간편결제'
df1.loc[df1.query('결제방법.str.startswith("후")').index, '결제방법'] = '후불'
df1.loc[df1.query('결제방법.str.startswith("정")').index, '결제방법'] = '정기결제'
# 결제방법은 맨 앞의 한가지만 사용한 것으로 간주
```

```python
df1.groupby(['결제방법'])['판매금액'].sum()
```

```python
df1 = df1.drop(df1[df1['결제방법'] == 'nan'].index) # 결측치 삭제
```

```python
df1.groupby(['결제방법']).agg({'판매금액': 'sum'})
```

```python
df1 = df1.reset_index(drop = True)
```

```python
df1_realmonth = df1.groupby(['월']).agg({'판매금액': 'sum'}).reset_index()
df1_realyear = df1.groupby(['년도']).agg({'판매금액': 'sum'}).reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(x='월', y='판매금액', data=df1_realmonth)
plt.title(f'월별 판매금액')
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(x='년도', y='판매금액',  data=df1_realyear)
plt.xticks([2019, 2020, 2021, 2022])
plt.title(f'년도별 판매금액')
plt.show()
```

```python
def format_sales_amount(amount):
    return '{:,.0f}원'.format(amount)
df1_realyear['판매금액'] = df1_realyear['판매금액'].apply(format_sales_amount)
```

```python
df1_realyear
```

```python
df1['할부개월수'] = df1.query('할부기간.notna() and 할부기간 != "1개월"')['할부기간'].apply(lambda x : x.split('개월')[0]).astype(int)
```

```python
df1['당월할부금'] = df1['판매금액']/df1['할부개월수']
```

```python
df1[df1['당월할부금'].notna()]
```

```python
df_monthly_revenue = df2.groupby('월')['판매금액'].sum() - df2.groupby('월')['당월할부금'].sum()
```

```python
df1_netprofit = df_monthly_revenue.to_frame()
```
