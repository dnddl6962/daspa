```python
import pandas as pd
from datetime import timedelta
```

```python
df = pd.read_excel('./미니프로젝트-쇼핑몰 실습데이터.xlsx')
```

```python
df
```

```python
df = df.query('처리상태 == "구매확정"')
```

```python
del df['초도상품']
```

```python
del df['제작문구 내역']
```

```python
df1 = df.copy()
```

```python
df1
```

```python
print(min(df1['주문일자']), max(df1['주문일자']))
```

```python
# 최대 주문일자 + 1일을 계산
snapshot_date = max(df1['주문일자']) + timedelta(days=1)
print(snapshot_date)
```

```python
datamart = df.groupby(['주문번호']).agg({'주문일자': lambda x: (snapshot_date - x.max()).days, '주문번호':'count', '판매금액':'sum'})
datamart.head()
```

```python
datamart.rename(columns = {'주문일자': 'Recency', '주문번호': 'Frequency', '판매금액': 'MonetaryValue'}, inplace=True)
datamart.head()
```

```python
r_labels = range(4, 0, -1)
r_quartiles = pd.qcut(datamart['Recency'], 4, labels = r_labels)
datamart = datamart.assign(R = r_quartiles)
datamart.head()
```

```python
datamart['Frequency'] = datamart['Frequency'].fillna(1)
```

```python
f_labels = (1, )
f_quartiles = pd.qcut(datamart['Frequency'], 4, labels=f_labels, duplicates='drop')
datamart = datamart.assign(F=f_quartiles.values)
```

```python
datamart
```

```python
datamart['F'].unique()
```

```python
datamart['MonetaryValue']
```

```python
m_labels = range(1, 5)
m_quartiles = pd.qcut(datamart['MonetaryValue'], 4, labels = m_labels)
datamart = datamart.assign(M = m_quartiles)
datamart.head()
```

```python
datamart['RFM_score'] = datamart[['R', 'F', 'M']].sum(axis=1)
datamart.groupby('RFM_score').agg({'Recency':'mean', 'Frequency': 'mean', 'MonetaryValue': ['mean','count']}).round(1)
```

```python
def medal_seg(df):
  if df['RFM_score'] >= 7:
    return 'A'
  elif (df['RFM_score'] >= 5) and (df['RFM_score'] < 7):
    return 'B'
  else:
    return 'C'

datamart['RFM_Grade'] = datamart.apply(medal_seg, axis=1)
```
